```python
import pandas as pd
```

```python
data_folder = '../datasets/'
data = pd.read_csv(f'{data_folder}marketing_campaign.csv' , sep=';')
data.head()
```

## Data cleaning, Feature Engineering

```python
data.info()
```

```python
data.describe()
```

```python
# Supprimer les colonnes Z_CostContact, Z_Revenue et ID
data = data.drop(columns=['Z_CostContact','Z_Revenue', 'ID'])
data.columns
```

```python
# Convertir la colonne Dt_Customer en date
data['Dt_Customer']=pd.to_datetime(data['Dt_Customer'])
data.info()
```

```python
data.isna().sum()
```

```python
# Ajouter une colonne Age
data['Age']=2024-data['Year_Birth']
# Supprimer la colonne Year_Birth
data = data.drop(columns='Year_Birth')
data.head()
```

```python
# Supprimer les lignes avec Income nulle
data = data.dropna(subset='Income')
data.shape
```

```python
data.isnull().sum()
```

```python
# Fusionner les colonnes KidHomme et TeenHome
data['Kid'] = data['Kidhome'] + data['Teenhome']
data = data.drop(columns=['Kidhome', 'Teenhome'])
data
```

```python
# Filtre de la colonne éducation
def education_level(education):
    if education in ["Graduation","phD","Master"]:
        return "high"
    elif education in ["Basic"]:
        return "Middle"
    else:
        return "Low"
    
data["Education_level"] = data["Education"].apply(education_level)

data['Marital_Status'].unique()
```

```python
# Filtre des situations matrimoniales
def living_status(marital_status):
    if marital_status in ['Alone', 'Absurd', 'YOLO', 'Single', 'Divorced']:
        return 'Alone'
    else:
        return 'Couple'

data['Living_Status'] = data['Marital_Status'].apply(living_status)
data.head()
```

```python
# Total de campagne accepté par le client
campagn_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
data['Total_Campaigns_Accepted'] = data[campagn_columns].sum(axis=1)
```

```python
# Moyenne des dépenses
spent_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
data['Average_Spend'] = (data[spent_columns].sum(axis=1)) / data['NumDealsPurchases']
```

```python
# Total des dépenses
data['Total_spent'] = data['MntWines']+data["MntWines"] +data['MntFruits']+ data['MntMeatProducts'] +data['MntFishProducts']+data['MntSweetProducts']+ data['MntGoldProds']
```

```python
data.describe()
```

```python
# il existe des ages > 100
data = data[(data['Age'] > 18) & (data['Age'] < 100)]
data.describe()
```

## Exportation des données clean

```python
# expotation des données clean
display(data.head())
data.to_csv(f'{data_folder}cleaned_data.csv', index=False)
```
