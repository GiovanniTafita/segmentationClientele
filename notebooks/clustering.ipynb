# importation 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

## importation fichier .csv

```python
data_path = '../datasets/cleaned_data.csv'
data = pd.read_csv(data_path )
```

# Préparation des données pour K-means

```python
features = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
X = data[features].fillna(0)
```

# Standardisation des données

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Application de K-means

```python
kmeans = KMeans(n_clusters=4, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)
```

# Visualisation des clusters

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Income'], y=data['MntWines'], hue=data['Cluster'], palette='Set1')
plt.title('Clustering des clients selon le revenu et les achats de vin')
plt.xlabel('Revenu')
plt.ylabel('Achats de vin')
plt.show()
```

# Taille des clusters

```python
cluster_sizes = data['Cluster'].value_counts()
print(cluster_sizes)
```
