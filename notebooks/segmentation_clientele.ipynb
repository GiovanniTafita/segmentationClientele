## Data cleaning

```python
# On voit que les colonnes Z_CostContact et Z_Revenue ont les même valeurs pour tous les individus
data = data.drop(columns=['Z_CostContact','Z_Revenue'])
data.head()
```

```python
# Suppression des lignes contenant des valeurs Icome null
data = data.dropna()

# Suppression de la colonne ID
data = data.drop(columns=['ID'])
data
```
